# VGR:s "Visa tider" för PCR
Letar upp alla listade kommuner, kollar alla listade mottagningar. De mottagningar som enbart länkar till 1177.se ska ha en "gul ruta" för att klara detta test.

## 1. Hämtar kommuner

```python
from bs4 import BeautifulSoup
import requests

startsida_pcr = 'https://www.vgregion.se/pcr-test-vgr/'
startsida_html = requests.get(startsida_pcr)
soup = BeautifulSoup(startsida_html.text, 'html.parser')

kommun_adresser = []

for link in soup.find_all('a'):
    if 'vgregion.se/ov/pcr-test/pcr-test/' in link.get('href'):
        kommun_adresser.append(link.get('href'))


print(f"Antal listade kommuner: {len(kommun_adresser)}")
print("Alla kommunadresser", kommun_adresser)
```

## 2. Hämtar alla mottagningar

```python
lankar_till_kontaktkort = []

for kommun in kommun_adresser:
    kommun_html = requests.get(kommun)
    soup = BeautifulSoup(kommun_html.text, 'html.parser')
    
    for link in soup.find_all('a'):
        if '/hitta-vard/kontaktkort/' in link.get('href'):
            lankar_till_kontaktkort.append(link.get('href'))
            print(link.get('href'))
    
```

## 3. Vilka har inte en gul ruta?

```python
print(f"har_gul_ruta;har_gul_ruta_och_bra_ord;vc_namn;vc_url")

ordlista = ['PCR-test', 'covid19', 'covid-19', 'SARS-CoV-2', 'antikropp', 'provtagning']
# göra lower()

def ordmarkaren(text):
    return_bool = 'nej'
    #print(type(text))
    
    for ord in ordlista:
        #print(ord.lower())

        if ord.lower() in text.lower():
            # print(ord.lower(), 'finns i sidans text')
            return_bool = 'ja'
    
    return return_bool

for vc_kontaktkort in lankar_till_kontaktkort:
    vc_html = requests.get(vc_kontaktkort)
    soup = BeautifulSoup(vc_html.text, 'html.parser')
    #print(vc_kontaktkort)
    
    if '<div class="contact-alert">' in vc_html.text:
        gul_text = soup.find("div", {"class": "contact-alert"})
        print(f"ja; {ordmarkaren(gul_text.getText())}; {soup.title.text}; {vc_kontaktkort}")
    else:
        print(f"nej; nej; {soup.title.text}; {vc_kontaktkort}")

```

## 4. Antal lediga tider
* Per kommun?

```python
import re

def hitta_antal_tider(kommun_text):
    r1 = re.findall(r"(?:\()(\d+)", kommun_text)
    tider = 0

    for mottagning_med_tider in r1:
        antal = int(mottagning_med_tider)
        tider += antal

    return tider

print(f"lediga_tider; kommun_namn")

for kommun_url in kommun_adresser:
    kommun_text = requests.get(kommun_url)
    soup = BeautifulSoup(kommun_text.text, 'html.parser')
    print(f"{hitta_antal_tider(soup.getText())}; {soup.title.text}")
```
