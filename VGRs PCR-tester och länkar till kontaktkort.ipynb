# VGR:s "Visa tider" för PCR
Letar upp alla listade kommuner, kollar alla listade mottagningar. De mottagningar som enbart länkar till 1177.se ska ha en "gul ruta" för att klara detta test.

## 1. Hämtar kommuner

```python
from bs4 import BeautifulSoup
import requests

startsida_pcr = 'https://www.vgregion.se/pcr-test-vgr/'
startsida_html = requests.get(startsida_pcr)
soup = BeautifulSoup(startsida_html.text, 'html.parser')

kommun_adresser = []

for link in soup.find_all('a'):
    if 'vgregion.se/ov/pcr-test/pcr-test/' in link.get('href'):
        kommun_adresser.append(link.get('href'))


print(f"Antal listade kommuner: {len(kommun_adresser)}")
print("Alla kommunadresser", kommun_adresser)
```

## 2. Hämtar alla mottagningar

```python
lankar_till_kontaktkort = []

for kommun in kommun_adresser:
    kommun_html = requests.get(kommun)
    soup = BeautifulSoup(kommun_html.text, 'html.parser')
    
    for link in soup.find_all('a'):
        if '/hitta-vard/kontaktkort/' in link.get('href'):
            lankar_till_kontaktkort.append(link.get('href'))
            print(link.get('href'))
    
```

## 3. Vilka har inte en gul ruta?

```python
print(f"har_gul_ruta;vc_namn;vc_url")

for vc_kontaktkort in lankar_till_kontaktkort:
    vc_html = requests.get(vc_kontaktkort)
    soup = BeautifulSoup(vc_html.text, 'html.parser')
    
    if '<div class="contact-alert">' in vc_html.text:
        print(f"ja;{soup.title.text};{vc_kontaktkort}")
    else:
        print(f"nej;{soup.title.text};{vc_kontaktkort}")

```
