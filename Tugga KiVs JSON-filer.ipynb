# Tugga KiVs JSON-filer

```python
import json
from tabulate import tabulate

f = open('data/20220519 units.json')
  
data = json.load(f)
f.close()       

resultat = []

## Check that there is a result 
for i in data['units']:
    dn = i['dn']
    hsaid = i['attributes']['hsaIdentity'][0]
    namn = i['attributes']['ou'][0] # ouShort finns inte alltid
    
    try:
        hsamngmtcode = i['attributes']['hsaManagementCode'][0] # om 4 är det en privat vårdgivare?
    except KeyError:
        hsamngmtcode = None
        
    try:
        kommunkod = i['attributes']['hsaMunicipalityCode'][0]
    except KeyError:
        kommunkod = None
    
    try:
        vgrAo3kod = i['attributes']['vgrAo3kod'][0]
    except KeyError:
        vgrAo3kod = None
    
    try:
        vgrObjectManagers = i['attributes']['vgrObjectManagers'][0]
    except KeyError:
        vgrObjectManagers = None
    
    try:
      hsaBusinessClassificationCode = i['attributes']['hsaBusinessClassificationCode'] # om "2029" finns i listan gör de covidvaccination
    except:
      hsaBusinessClassificationCode = None
    
    
    #if 'hemma' in namn.lower() and 'ou=vårdval' in dn.lower():
    #if 'ou=vårdval' in dn.lower():
    #if hsaBusinessClassificationCode != None and '2029' in hsaBusinessClassificationCode: # covidvaccination
    #if hsaBusinessClassificationCode != None and '1519' in hsaBusinessClassificationCode: # ungdomsmottagning
    #if hsaBusinessClassificationCode != None and '1504' in hsaBusinessClassificationCode: # 1504;Barnhälsovård/BVC
    if hsaBusinessClassificationCode != None and '1516' in hsaBusinessClassificationCode and '1402' in hsaBusinessClassificationCode and '1413' in hsaBusinessClassificationCode: # 1402, 1413, 1516
        resultat.append([hsamngmtcode, vgrAo3kod, namn, vgrObjectManagers, hsaid])


print(tabulate(resultat, headers=['hsamngmtcode', 'vgrAo3kod', 'namn', 'vgrObjectManagers', 'hsaid']))
```
